```python
master_results_00 = results_df.copy()
```

```python
master_results_01 = pd.DataFrame(master_results_00, columns = ["resultId", "raceId", "driverId", "constructorId", "position", "laps", "milliseconds", "fastestLap", "fastestLapSpeed"])

master_results_00.info()
```

```python
master_results_01
```

```python
master_results_01["position"] = master_results_01["position"].replace("\\N", pd.NA)
master_results_01["milliseconds"] = master_results_01["milliseconds"].replace("\\N", pd.NA)
master_results_01["fastestLap"] = master_results_01["fastestLap"].replace("\\N", pd.NA)
master_results_01["fastestLapSpeed"] = master_results_01["fastestLapSpeed"].replace("\\N", pd.NA)

master_results_01["position"] = pd.to_numeric(master_results_01["position"], errors = "coerce").astype("Int64")
master_results_01["milliseconds"] = pd.to_numeric(master_results_01["milliseconds"], errors = "coerce").astype("Int64")
master_results_01["fastestLap"] = pd.to_numeric(master_results_01["fastestLap"], errors = "coerce").astype("Int64")
master_results_01["fastestLapSpeed"] = pd.to_numeric(master_results_01["fastestLapSpeed"], errors = "coerce").astype(float)
```

```python
master_results_01.count()
```

```python
master_results_01.info()
```

```python
master_results_01["finishedRace"] = np.where(master_results_01["milliseconds"].notnull(), 1, 0)
master_results_01[master_results_01["milliseconds"].isnull()].head()
```

```python
master_results_02 = pd.merge(master_results_01, constructors_df[["constructorId", "name"]], on = "constructorId", how = "left")
```

```python
master_results_02.rename(columns = {"name": "constructorName"}, inplace = True)
master_results_02
```

```python
master_results_03 = pd.merge(master_results_02, drivers_df[["driverId", "forename", "surname"]], on = "driverId", how = "left")
master_results_03
```

```python
master_results_03["driverName"] = master_results_03["forename"] + " " + master_results_03["surname"]
master_results_03.drop("forename", axis = 1, inplace = True)
master_results_03.drop("surname", axis = 1, inplace = True)
master_results_03
```

```python
master_results_04 = pd.merge(master_results_03, lap_times_df[["raceId", "driverId", "lap", "milliseconds"]], left_on = ["raceId", "driverId", "fastestLap"], right_on = ["raceId", "driverId", "lap"], how = "left")
master_results_04
```

```python
master_results_04.drop("lap", axis = 1, inplace = True)
master_results_04
```

```python
master_results_04.rename(columns = {"milliseconds_x": "overallTimeMsec", "milliseconds_y": "fastestLapTimeMsec"}, inplace = True)
master_results_04
```

```python
master_results_04["fastestLapTimeMsec"] = pd.to_numeric(master_results_04["fastestLapTimeMsec"], errors = "coerce").astype("Int64")
master_results_04.head()
```

```python
master_results_04.drop("fastestLap", axis = 1, inplace = True)
master_results_04.head()
```

```python
circuits_df.head()
```

```python
races_df.head()
```

```python
master_results_05 = pd.merge(master_results_04, races_df[["raceId", "date"]], on = "raceId", how = "left")
```

```python
master_results_05.head()
```

```python
master_results_05["date"] = pd.to_datetime(master_results_05["date"])
master_results_05["year"] = master_results_05["date"].dt.year
master_results_05["month"] = master_results_05["date"].dt.month
master_results_05["day"] = master_results_05["date"].dt.day

master_results_05.head()
```

```python
master_results_06 = master_results_05[["resultId", "raceId", "driverId", "constructorId", "driverName", "constructorName", "overallTimeMsec", "fastestLapTimeMsec", "fastestLapSpeed", "position", "laps", "finishedRace", "date", "year", "month", "day"]]
master_results_06.head()
```

```python
master_results_06.info()
```

```python
master_results_06["position"] = master_results_06["position"].astype("float64")
```

```python
master_results_06["win"] = np.where((master_results_06["position"] == 1), 1, 0)
```

```python
master_results_06.head()
```

```python
master_results_07 = master_results_06.copy()
```

```python
master_results_07["tbl"] = (master_results_07["overallTimeMsec"] - master_results_07.groupby("raceId")["overallTimeMsec"].transform("min"))
```

```python
master_results_07.head()
```

```python
master_results_08 = master_results_07[["resultId", "raceId", "driverId", "constructorId", "driverName", "constructorName", "tbl", "overallTimeMsec", "fastestLapTimeMsec", "fastestLapSpeed", "position", "win", "laps", "finishedRace", "date", "year", "month", "day"]]

master_results_08.head()
```
